```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

import freud # для создания диаграммы Вороного
```

```python
eps = 0.03
```

```python
points = np.asarray([
                    [0.3, 0.1, 0], 
                    [0.3, -0.1, 0],
                    [-0.3, 0.1, 0],
                    [-0.3, -0.1, 0],
                    [0.1, 0, 0],
                    [-0.1, 0, 0],
                    [0.5+eps, 0, 0],
                    [0, 0.5, 0],
                    [0.5+eps, 0.5, 0]
                    ]) # исправить чтоб было как на картинке
```

```python
box = freud.box.Box(Lx=1+2*eps, Ly=1, is2D=True)
points = box.wrap(points)
```

```python
voro = freud.locality.Voronoi()
voro.compute((box, points))

plt.figure()
ax = plt.gca()
voro.plot(ax=ax, cmap="RdBu")
ax.set_xlim(-2, 2)
ax.set_ylim(-1, 1)
ax.scatter(points[:, 0], points[:, 1], s=2, c="k")
plt.show()
```
