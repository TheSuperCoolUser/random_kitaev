```python
N = 5
stats = {}
graphs = []
for i in range(300): # 
    G = voronoi_adjacency_matrix(N, return_nx_graph = True, draw = False)
    if G == None:
        continue
    
    seen = False
    for g in graphs:
        if nx.is_isomorphic(g, G):
            seen = True
            stats[g] = stats.get(g, 0) + 1
        else:
            continue
    if (not seen):
        graphs.append(G)
        stats[G] = stats.get(G, 1)
```
